```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import csv
import re
from selenium.common.exceptions import NoAlertPresentException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

### class_dada

```python

driver = webdriver.Chrome()

# 강의 목록 페이지로 이동
driver.get("https://www.naksinuri.kr/lesson/class/list.do")

wait = WebDriverWait(driver, 10)

li = wait.until(EC.element_to_be_clickable((By.XPATH, '//li[@onclick="fnSelectInfs3(\'바다\')"]')))
li.click()
time.sleep(2) 

a = wait.until(EC.element_to_be_clickable((By.XPATH, '//a[@onclick="fnSelectInfs(4); return false;"]')))
a.click()
time.sleep(2) 

# 'onclick' 속성을 가진 span 요소들 추출
span = driver.find_element(By.XPATH, f'//span[@onclick="view2(\'159\')"]')
span.click()
time.sleep(2)


with open("class_data.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["제목", "내용"])  # 헤더 작성
    while True:
        soup = BeautifulSoup(driver.page_source, "html.parser")
        title = soup.find("h3", class_="tit")
        content = soup.find("div", class_="content")
        
        title_text = title.text.strip() if title else "제목 없음"
        content_text = content.text.strip() if content else "내용 없음"
        print(f"[{title_text}]\n{content_text[:100]}...\n")

        writer.writerow([title_text,content_text])
        try:
            # 이전 버튼 클릭
            prev_btn = driver.find_element(By.XPATH, '//a[@onclick and contains(@onclick, "next")]')
            prev_btn.click()
            time.sleep(2)

            # 경고창(Alert) 뜨면 종료
            try:
                alert = driver.switch_to.alert
                print(f"경고창 내용: {alert.text}")
                alert.accept()
                print("더 이상 이전 강의 없음. 종료.")
                break
            except NoAlertPresentException:
                pass  # 경고창 없으면 계속 진행

        except Exception as e:
            print(f"예상치 못한 오류: {e}")
            break

driver.quit()
```

### binding_data

```python
driver = webdriver.Chrome()

# 강의 목록 페이지로 이동
driver.get("https://www.naksinuri.kr/lesson/binding/list.do")
time.sleep(2)  # 페이지 로드 대기

# 'onclick' 속성을 가진 span 요소들 추출
spans = driver.find_elements(By.XPATH, '//span[@onclick]')

lesson_ids = []
for span in spans:
    onclick = span.get_attribute('onclick')
    match = re.search(r"view2\('(\d+)'\)", onclick)
    if match:
        lesson_ids.append(match.group(1))

print(f"강의 ID 수: {len(lesson_ids)}")


with open("binding_data.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["강의 ID", "제목", "내용"])  # 헤더 작성

    for lesson_id in lesson_ids:
        driver.get("https://www.naksinuri.kr/lesson/binding/list.do")
        time.sleep(2)

        # 다시 span 요소 찾아서 해당 ID 클릭
        span = driver.find_element(By.XPATH, f'//span[@onclick="view2(\'{lesson_id}\')"]')
        span.click()
        time.sleep(2)  # 페이지가 로드되도록 대기

        soup = BeautifulSoup(driver.page_source, "html.parser")
        title = soup.find("h3", class_="tit")
        content = soup.find("div", class_="content")
        p_tags = content.find_all('p')
        for p in p_tags:
            img = p.find('img')
            if img:
                img_tag = img
                break 
        alt_text = img_tag.get('alt') if img_tag and img_tag.get('alt') else ""


        title_text = title.text.strip() if title else "제목 없음"
        print(f"[{title_text}]\n{alt_text[:100]}...\n")

        writer.writerow([lesson_id, title_text,alt_text])

driver.quit()
```
